### Basic Graphs

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.figure(figsize=(6,4),dpi=100)
plt.plot(x,y, label='2x' ,color='r', linewidth=2,
         marker='.',markersize=15,markeredgecolor='b',markerfacecolor='g',
         linestyle=':')
plt.title("Graph 1",loc='left')
plt.xlabel('x axis!')
plt.ylabel('y axis!')
plt.grid(axis='x',color='hotpink',ls='--',lw=.5)
plt.legend()
plt.savefig('faa graph.png', dpi=300, bbox_inches='tight', transparent=False)
plt.show()
```

```python
#plotting without lines
plt.plot(x,y,'o')
plt.show()
```

```python
x2=np.arange(0,4.5,.5)
# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')
plt.show
```

### Object-Oriented Programming (OOP) in Python combined with Matplotlib

```python
import matplotlib.pyplot as plt
class Point:
	def __init__(self, x,y, color="red", size=10):
		self.x = x
		self.y = y
		self.color = color
		self.size = size

	def __add__(self, other):
		if isinstance(other, Point):
			x = self.x + other.x
			y = self.y + other.y
			return Point(x,y)
		else:
			x = self.x + other
			y = self.y + other
			return Point(x, y, color=self.color)

	def plot(self):
		plt.scatter(self.x, self.y, c=self.color, s=self.size)

if __name__ == "__main__":
	a = Point(8,3)
	b = Point(2,4)
	c = a + b
	c.plot()
	plt.show()

# It specifically showcases Operator Overloading...
```
